# 05 - Conclusiones

**Objetivo del proyecto:** Analizar cómo se relaciona el tiempo mensual de visualización con el tipo de plan de suscripción y el país de los usuarios, para identificar patrones de consumo en la plataforma de streaming.

Esta notebook integra los hallazgos de `03_eda.ipynb` y `04_pca.ipynb`, diferenciando **evidencia**, **interpretación** y **conclusión**, y debe mantenerse coherente con el informe final (`reports/informe_final.pdf`) y la aplicación Streamlit.

## 1. Respuesta a los objetivos del análisis

**Evidencia:**
- La mediana de `monthly_watch_time_mins` por plan es: Básico 552.9 min, Estándar 839.1 min, Premium 1121.7 min (notebook 03).
- La mediana de `monthly_watch_time_mins` por país varía solo entre 757.7 min (México) y 795.0 min (Uruguay) (notebook 03).
- El salto de consumo al subir de plan (Básico -> Premium) es de entre 540 y 610 minutos, y esa magnitud es consistente en los 7 países (notebook 03, heatmap).
- En PCA, las 3 variables numéricas (`age`, `monthly_watch_time_mins`, `customer_support_tickets`) tienen correlaciones cercanas a cero entre sí, y ninguna domina las primeras componentes (notebook 04).
- La proyección PCA no separa a los usuarios por plan, porque el efecto del plan es específico de `monthly_watch_time_mins` y no se refleja en `age` ni en `customer_support_tickets` (notebook 04).

**Interpretación:** el plan de suscripción está fuertemente asociado al tiempo de visualización -- un usuario Premium consume, en mediana, el doble de minutos que uno Básico. El país, en cambio, introduce una variación mucho más chica (~5%) y de forma pareja en todos los planes, sin que ningún país se destaque de forma desproporcionada dentro de un mismo plan.

**Conclusión:** respondiendo al objetivo del proyecto, el patrón de consumo en la plataforma está **principalmente explicado por el plan de suscripción**, mientras que el **país actúa como un factor secundario y consistente**, no como un modificador fuerte del efecto del plan. No se observa una interacción marcada entre ambas variables: subir de plan produce un incremento de consumo de magnitud similar sin importar el país.

## 2. Limitaciones

- El dataset no contiene información sobre el contenido consumido (catálogo, tipo de dispositivo, franja horaria), que podría explicar mejor las diferencias de consumo dentro de un mismo plan.
- Varios valores de `monthly_watch_time_mins` fueron imputados con la mediana por (plan, país) durante la limpieza (262 registros, notebook 02) por corresponder a datos faltantes o inválidos en el dataset original; esto puede suavizar levemente la variabilidad real dentro de cada grupo.
- El alcance de las conclusiones se encuentra condicionado por la información disponible y por las decisiones documentadas durante el proceso de limpieza.

## 3. Mejoras futuras

- Una mejora futura podría consistir en incorporar información adicional (tipo de dispositivo, género consumido con mayor detalle, franja horaria de uso) que permita ampliar el alcance del análisis y explicar la variabilidad de consumo dentro de un mismo plan.
- Podría analizarse la evolución temporal del consumo usando `last_login_date` una vez completada esa información, para detectar tendencias de uso más allá de la foto estática actual del dataset.
